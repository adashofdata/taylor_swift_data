## Step 1: Read in Text

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

pd.set_option('display.max_colwidth', None)
```

```python
# Read in lyrics extracted from the Genius API
data = pd.read_csv('../Taylor_Swift_Genius/taylor_swift_genius_data.csv')
data.head()
```

## Step 2: Clean Text

```python
def clean_text(text):
    
    # Some additional light data cleaning - you will need to adjust based on your data    
    text = text.replace('you might also like',' ') # found this hidden in the text
    
    return text

# Apply an extra round of text cleaning - you can apply more, if needed for your data
cleaned_text = pd.DataFrame(data['Lyrics'].apply(lambda x: clean_text(str(x))))
cleaned_text["Song Name"] = data["Song Name"]

# Rearrange the columns
cleaned_text = cleaned_text[['Song Name', 'Lyrics']]
cleaned_text
```

## Step 3: Split Lyrics into Words

```python
# Use CountVectorizer to turn the DataFrame into a document-term matrix
cv = CountVectorizer() # By default, turns text into lowercase and removes English stop words

data_cv = cv.fit_transform(cleaned_text.iloc[:,1])
data_dtm_cv = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm_cv
```

```python
ts_words = pd.concat([cleaned_text["Song Name"], data_dtm_cv], axis=1)
ts_words
```

## Step 4: Export the Words as a CSV File

```python
ts_words.to_csv("taylor_swift_words_data.csv", index=False)
```
